# Energy Consumption Data Visualization
This notebook visualizes the electricity consumption data for Network Levels 5 and 7 (NE5, NE7) in the city of Zurich.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from IPython.display import display
```

## 1. Load Data
Load the CSV data and parse the Timestamp column to be a datetime index.

```python
# Load data
df = pd.read_csv('Data/ewz_stromabgabe_netzebenen_stadt_zuerich.csv')

# Convert timestamp string to datetime objects
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Set timestamp as index and sort it
df = df.set_index('Timestamp').sort_index()

print("Data shape:", df.shape)
display(df.head())
```

## 2. Interactive Time-Series Visualization
Let's visualize the raw data dynamically.

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df['Value_NE5'], mode='lines', name='NE5 Consumption', line=dict(color='#1f77b4')))
fig.add_trace(go.Scatter(x=df.index, y=df['Value_NE7'], mode='lines', name='NE7 Consumption', line=dict(color='#ff7f0e')))

fig.update_layout(
    title="Electricity Consumption (NE5 & NE7) Over Time",
    xaxis_title="Time",
    yaxis_title="Electricity Consumption (kWh)",
    hovermode="x unified",
    template="plotly_white"
)

fig.show()
```

## 3. Data Resampling & Aggregation
Because the dataset is very large, it's often more insightful to aggregate by Day, Week, or Month.

```python
# Resample data by different time units
df_daily = df.resample('D').sum()
df_weekly = df.resample('W').sum()
df_monthly = df.resample('ME').sum()

# Plot Monthly totals as an example
fig_monthly = go.Figure()
fig_monthly.add_trace(go.Bar(x=df_monthly.index, y=df_monthly['Value_NE5'], name='NE5 (Monthly)', marker_color='#1f77b4'))
fig_monthly.add_trace(go.Bar(x=df_monthly.index, y=df_monthly['Value_NE7'], name='NE7 (Monthly)', marker_color='#ff7f0e'))

fig_monthly.update_layout(
    title="Monthly Total Electricity Consumption",
    xaxis_title="Month",
    yaxis_title="Total Electricity Consumption (kWh)",
    barmode='group',
    template="plotly_white"
)

fig_monthly.show()
```
